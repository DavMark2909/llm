```python
import re

def sql_query_extraction(query):
    pattern = r'(INSERT.*?;)'
    matches = re.findall(pattern, query, re.DOTALL)
    if not matches:
        raise ValueError("No INSERT queries found in the provided text.")
    queries = []
    for sql in matches:
        queries.append(sql)
    return queries

def table_query_creation(counter):
    record = test_schemas[counter]
    queries = []
    for key, value in record.items():
        values = value.split(",")
        cleaned = [element.replace('\n', '').strip() for element in values]
        query = data["etl"]["utils"]["table_creation"].format(table_name=key, table_content=", ".join(cleaned))
        queries.append(query)
    return queries

def table_content_insertion(counter):
    content = ",".join(table_query_creation(counter))
    model_prompt = data["etl"]["prompt"].format(table_schemas=content)
    response = llm.invoke(model_prompt)
    try:
        queries = sql_query_extraction(response.content)
        for sql in queries:
            cursor.execute(sql)
            conn.commit()
    except ValueError as e:
        raise e
    
def table_content_extraction(counter):
    tables = test_schemas[counter]
    for key in tables.keys():
        query = data["etl"]["utils"]["table_content_extraction"].format(table_name=key)
        res = cursor.execute(query)
        print('From a table ' + key)
        print(res.fetchall())

def table_deletion(counter):
    record = test_schemas[counter]
    queries = []
    for key in record.keys():
        queries.append(data["etl"]["utils"]["table_deletion"].format(table_name=key))
    queries.append(data["etl"]["utils"]["table_deletion"].format(table_name=table_name_extraction(counter)))
    for sql in queries:
        cursor.execute(sql)
        conn.commit()

def etl_job_generation(counter):
    res = llm.invoke(data["etl"]["role"].format(sql_context="sqlite", tables_schema=test_schemas[counter], generated_fact_table=generated_fact_schemas[counter]))
    a = sql_query_extraction(res.content)
    print(a[0])
    try:
        cursor.execute(a[0])
        conn.commit()
        table_name = table_name_extraction(counter)
        query = data["etl"]["utils"]["table_content_extraction"].format(table_name=table_name)
        res = cursor.execute(query)
        print('From a table ' + table_name)
        print(res.fetchall())
    except:
        raise ValueError("Error while trying to execute the ETL job.")

def table_name_extraction(counter):
    string = generated_fact_schemas[counter]
    splitted = string.split("\n")
    return splitted[1].strip().split(" ")[2]


def testing_loop(counter):
    tables = table_query_creation(counter)
    for table in tables:
        cursor.execute(table)
        conn.commit()
    print("Created tables")
    try:
        table_content_insertion(counter)
        print("inserted elements")
    except ValueError as e:
        print(e)
        return -1
    table_content_extraction(counter)
    cursor.execute(generated_fact_schemas[counter])
    conn.commit()
    try:
        etl_job_generation(counter)
        table_deletion(counter)
        return 1
    except ValueError as e:
        print(e)
        table_deletion(counter)
        return -1
    
def testing(j=0):
    correct = 0
    wrong = 0
    for i in range(j, len(test_schemas)):
        print(i)
        if testing_loop(i) < 0:
            wrong += 1
        else:
            correct += 1
    total = correct + wrong
    return (correct / total, wrong / total)
```

```python
from typing_extensions import TypedDict

class EtlTestingState(TypedDict):
    test_count: int
    table_queries: list[str]
    generated_etl: str
    etl_output: str
    errors: str
    fact_table_generated: bool
    retries: int
```

```python
# helping functions---------------------------
def sql_query_extraction(query):
    pattern = r'(INSERT.*?;)'
    matches = re.findall(pattern, query, re.DOTALL)
    if not matches:
        raise ValueError("No INSERT queries found in the provided text.")
    queries = []
    for sql in matches:
        queries.append(sql)
    return queries

def table_name_extraction(counter):
    string = generated_fact_schemas[counter]
    splitted = string.split("\n")
    return splitted[1].strip().split(" ")[2]
# --------------------------------------------

#routing functions: --------------------------
def error_table_creation_checker(state: EtlTestingState):
    if state['errors'] == "":
        return "continue"
    else:
        if state['fact_table_generated']:
            return "fact_table_deletion"
        else:
            return "init_table_deletion"

def error_etl_generation_checker(state: EtlTestingState):
    if state["errors"] == "":
        return "continue"
    else:
        if state['retries'] < 3:
            return "repeat"
        else:
            return "continue"
# --------------------------------------------

#tools
def table_query_creation_state(state: EtlTestingState):
    record = test_schemas[state["test_count"]]
    queries = []
    for key, value in record.items():
        values = value.split(",")
        cleaned = [element.replace('\n', '').strip() for element in values]
        query = data["etl"]["utils"]["table_creation"].format(table_name=key, table_content=", ".join(cleaned))
        queries.append(query)
    state['table_queries'] = queries
    return state

def table_creator_state(state: EtlTestingState):
    tables = state['table_queries']
    for table in tables:
        cursor.execute(table)
        conn.commit()
    return state

def table_content_insertion_state(state: EtlTestingState):
    content = ",".join(table_query_creation(state['test_count']))
    model_prompt = data["etl"]["prompt"].format(table_schemas=content)
    response = llm.invoke(model_prompt)
    try:
        queries = sql_query_extraction(response.content)
        for sql in queries:
            cursor.execute(sql)
            conn.commit()
    except ValueError as e:
        state['errors'] = str(e)
    return state
    
def table_content_extraction_state(state: EtlTestingState):
    tables = test_schemas[state['test_count']]
    for key in tables.keys():
        query = data["etl"]["utils"]["table_content_extraction"].format(table_name=key)
        cursor.execute(query)
    return state

def fact_table_generator_state(state: EtlTestingState):
    cursor.execute(generated_fact_schemas[state["test_count"]])
    state["fact_table_generated"] = True
    return state

def etl_job_generation_state(state: EtlTestingState):
    counter = state['test_count']
    res = llm.invoke(data["etl"]["role"].format(sql_context="sqlite", tables_schema=test_schemas[counter], generated_fact_table=generated_fact_schemas[counter]))
    a = sql_query_extraction(res.content)
    state["generated_etl"] = a[0]
    try:
        cursor.execute(a[0])
        conn.commit()
        table_name = table_name_extraction(counter)
        query = data["etl"]["utils"]["table_content_extraction"].format(table_name=table_name)
        res = cursor.execute(query)
        state["etl_output"] = res.fetchall()
        state["errors"] = ""
        return state
    except:
        state["errors"] = "Error while trying to execute the ETL job."
        state["retries"] += 1
    return state
    
def all_table_deletion_state(state: EtlTestingState):
    counter = state["test_count"]
    record = test_schemas[counter]
    queries = []
    for key in record.keys():
        queries.append(data["etl"]["utils"]["table_deletion"].format(table_name=key))
    queries.append(data["etl"]["utils"]["table_deletion"].format(table_name=table_name_extraction(counter)))
    for sql in queries:
        cursor.execute(sql)
        conn.commit()
    return state

def init_table_deletion_state(state: EtlTestingState):
    counter = state["test_count"]
    record = test_schemas[counter]
    queries = []
    for key in record.keys():
        queries.append(data["etl"]["utils"]["table_deletion"].format(table_name=key))
    for sql in queries:
        cursor.execute(sql)
        conn.commit()
    return state
    
```

```python
from langgraph.graph import StateGraph, START, END

graph = StateGraph(EtlTestingState)
graph.add_node("tables_extractor", table_query_creation_state)
graph.add_node("table_creator", table_creator_state)
graph.add_node("table_inserter", table_content_insertion_state)
graph.add_node("table_content_extractor", table_content_extraction_state)
graph.add_node("fact_table_generator", fact_table_generator_state)
graph.add_node("etl_job_generator", etl_job_generation_state)
graph.add_node("init_table_deletion", init_table_deletion_state)
graph.add_node("all_table_deletion", all_table_deletion_state)

graph.add_edge(START, "tables_extractor")
graph.add_edge("tables_extractor", "table_creator")
graph.add_edge("table_creator", "table_inserter")
graph.add_conditional_edges("table_inserter",
    error_table_creation_checker,
    {
        "continue": "table_content_extractor",
        "fact_table_deletion": "all_table_deletion",
        "init_table_deletion": "init_table_deletion"
    }
)
graph.add_edge("table_content_extractor", "fact_table_generator")
graph.add_edge("fact_table_generator", "etl_job_generator")
graph.add_conditional_edges("etl_job_generator", 
    error_etl_generation_checker,
    {
        "continue": "all_table_deletion",
        "repeat": "etl_job_generator"
    }
)
graph.add_edge("init_table_deletion", END)
graph.add_edge("all_table_deletion", END)

app = graph.compile()
```
