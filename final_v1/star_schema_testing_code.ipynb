```python
from typing_extensions import TypedDict

class TestingState(TypedDict):
    test_count: int
    user_input: str
    model_output: str
    model_feedback: str
    loops_count: int
```

```python
def generate_test_prompt_testing(state: TestingStateV2):
    counter = state["test_count"]
    prompts = data["model"]["prompts"]
    time_aggregation_string = ""

    agg_columns = test_agg_columns[counter]
    t_columns = test_columns[counter]
    operations = test_operations[counter]

    role_prompt = prompts["role"]
    tables_schemas_prompt = prompts["tables"].format(tables=generate_table_schema(counter))
    operations_string = ", ".join(f"{op} of {col}" for op, col in zip(operations, t_columns))
    operations_prompt = prompts["operations"].format(operations=operations_string)

    if test_time_aggregation[counter] != "NONE":
        generated_time_field = generate_time_aggregation(agg_columns[-1], test_time_aggregation[counter])
        time_aggregation_string = prompts["time_aggregation"].format(date_field=agg_columns[-1],time=test_time_aggregation[counter], generated_time_name=generated_time_field)
        agg_columns[-1] = generated_time_field
    if len(agg_columns) > 1:
        output_prompt = prompts["remarks_multiple_aggregation"].format(key={', '.join(agg_columns)})
    else:
        output_prompt = prompts["remarks_one_aggregation"].format(primary_key=agg_columns[0])

    user_prompt = prompts["user_prompt"].format(requirements=test_user_inputs[counter])
    a = ' '.join([role_prompt, tables_schemas_prompt, operations_prompt, time_aggregation_string, user_prompt, output_prompt])
    response = llm.invoke(a)
    result = response.content
    print("First generation of the response")
    print(result)
    state["model_output"] = result
    state["user_input"] = test_user_inputs[counter]
    return state

    
```

```python
def output_checker(state: TestingState):
    model_output = state["model_output"]
    user_input = state["user_input"]

    prompt = data["model"]["prompts"]["checking_prompt"].format(user_prompt=user_input, fact_table=model_output)
    res = llm.invoke(prompt)
    
    state["model_feedback"] = res.content
    return state

def output_checker_router(state: TestingState):
    res = state["model_feedback"]
    if res.lower() != "yes":
        # state["loops_count"] += 1
        return "repeat_fact_table_generation"
    else:
        return "fact_table_generated_succesfully"
    
def output_fixer(state: TestingState):
    prompt = data["model"]["prompts"]["fixing_prompt"].format(user_prompt=state["user_input"], feedback=state["model_feedback"], fact_table=state["model_output"])
    response = llm.invoke(prompt)
    state["model_output"] = response.content
    return state
```

```python
def output_checker_v2(state: TestingStateV2):
    model_output = state["model_output"]
    user_input = state["user_input"]

    prompt = data["model"]["prompts"]["checking_prompt"].format(user_prompt=user_input, fact_table=model_output)
    res = llm.invoke(prompt)
    state["model_output"] = res.content
    return state
```

```python
from langgraph.graph import StateGraph, START, END

graph = StateGraph(TestingState)
graph.add_node("generate_prompt", generate_test_prompt_testing)
graph.add_node("model_output_checker", output_checker)
graph.add_node("model_output_fixer", output_fixer)


graph.add_edge(START, "generate_prompt")
graph.add_edge("generate_prompt", "model_output_checker")
graph.add_edge("model_output_fixer", "model_output_checker")
graph.add_conditional_edges("model_output_checker",
    output_checker_router,
    {
        "repeat_fact_table_generation": "model_output_fixer",
        "fact_table_generated_succesfully": END
    })

app = graph.compile()
```
